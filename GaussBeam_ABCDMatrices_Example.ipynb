## (1) Define an initial Gaussian beam. Provide W1, R1 and calculate q1. ##

```python
lam = 700e-9 * 1000    # wavelength in mm
R1  = -2               # Radius of curvature in mm
W1  = 0.5              # Waist in mm
z1  = -1               # Position in mm   
```

### Calculate q1 using W1, R1

```python
invQ = 1/R1 - 1j *lam / (np.pi*W1**2)
q1   = 1/invQ
```

## (2) Define the ABCD matrix. Here, a free space propagation by distance d.

```python
d  = 0.2                            # distnace d in mm    
M  = np.array( [[1,d],[0,1]] )      # see slides for matrix
```

Apply the matrix several times. Propagate the beam forward several and store the resulting z-position and q-parameters in vector qVect and zVect

```python
# initialize position and parameter vectors
qVect  = [q1]
zVect  = [z1]
```

## (3) Calculate the new q-parameters as you propagate the beam forward several times.

```python
# initialize position and parameter vectors
qVect  = [q1]
zVect  = [z1]
```

```python
for i1 in range(11):
    # "multiply" matrix with q1
    # use application rule
    q1 = qVect[-1]
    z1 = zVect[-1]
    
    q2 = PH455.multiBeamMat(M, q1)
    z2 = z1 + d
    
    qVect.append( q2 )
    zVect.append( z2)
```
