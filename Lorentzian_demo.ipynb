# Lorentzian function. 
This is a demonstration file for the course PH455/957. The script demonstrates how to plot a Lorentzian function.

```python
import numpy as np                     # numpy arrays
import matplotlib.pyplot as plt        # plotting
import PH455
```

### Define x vector and parameters

```python
# set an x-range for plotting
x = np.linspace(-10,10,20001)

# set paramters for the Lorentzian
x0 = 0
b  = 1
```

### Plot Lorentzian and Gaussian function

```python
# plot a Lorentzian
plt.plot(x, PH455.Lorentzian(x,x0,b), 'b-' )

# plot a Gaussian with the same peak value
#plt.plot(x, PH455.Gaussian(x, 0, 2*b*0.6266 ), 'r-')

# plot a Gaussian with the same width
plt.plot(x, PH455.Gaussian(x, 0, b), 'r--' )

plt.xlabel('x-values')
plt.ylabel('y-values')
```
