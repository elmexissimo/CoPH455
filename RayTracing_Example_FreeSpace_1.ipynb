## Single ray tracing step ##
This is a demonstration file for PH455/957. It shows you how to implement a single ray tracing step.

```python
import numpy as np
import matplotlib.pyplot as plt
import PH455
```

(1) Define input ray: [y theta]

```python
rayIn    = np.array( [2, 0.2] )    # ray at height 2mm and angle 0.2 rad
zIn      = 0                       # for position 
```

(2) Define ray-transfer matrix for free space propagation over distance d

```python
d  = 5                             # 5 mm    
M  = np.array( [[1,d],[0,1]] )     # see slides for matrix
```

Calculate the output ray with M and rayIn

```python
rayOut = np.matmul(M, rayIn)       # multiply ray with matrix    
zOut   = zIn + d                   # new ray position
```

(3) Plot the ray 
"showRay( zList, rayList )" is a simple plotting function in the support module PH455.

- vertical gray lines - provided z-positions
- horizontal gray line - optical axis
- red circles - provided y-positions of rays
- dashed red lines - indicate ray between two points
- dashed blue lines - indicate angle of input and output rays

```python
PH455.showRay( [zIn,zOut], [rayIn, rayOut] )
```
