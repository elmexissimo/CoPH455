# Multiple interference for M waves #
This is a demonstration file for PH455/957. The file demonstrates the interference for M waves with increasing phase phi. 

```python
import numpy as np
import matplotlib.pyplot as plt
import PH455
```

Function to calculate the complex amplitude Um of a single wave.

```python
def Um(m,phi):
    I0 = 1
    h = 1
    return np.sqrt(I0)*h**(m-1)*np.exp(1j*(m-1)*phi)
```

Function to calculate the sum of several waves. Sum over the terms Um, starting with m=1 to M.

```python
def sumU(M,phi):
    U = 0
    for m in range(1,M+1):
        U = U + Um(m,phi)
        
    return U
```

Initialize M and phi

```python
M   = 8
phi = 0.3*np.pi
```

## Plot a phasor diagram for M waves ##

```python
PH455.showPhasor(sumU, phi, M)
```

## Show the total intensity of the waves vs. phi ##

Define an array of phase values and calculate the total intensity of the waves for each phase. The intensity is given by the square of the total U value.

```python
phiVect = np.linspace(0,5*np.pi,1000)
Intensity = []

for p in phiVect:
    Intensity.append( np.abs(sumU(M,p))**2 )
```

Plot the intensity vs. phiVect

```python
plt.plot(phiVect/(2*np.pi),Intensity)
plt.xlabel('Phase phi')
plt.ylabel('Intensity (arb.)')
plt.title( f"Intensity for M={M}" )
plt.show()
```

Plot the analytical function for the total intensity of M waves of equal amplitude.

```python
np.seterr(invalid='ignore') # Suppress the 1/0 warning

IntAnalytical = []
for p in phiVect:
    Int = np.sin(M*p/2)**2 / np.sin(p/2)**2
    IntAnalytical.append( Int )

plt.plot(phiVect/(2*np.pi),Intensity, 'b')
plt.plot(phiVect/(2*np.pi), IntAnalytical, '--r')
plt.xlabel('Phase phi')
plt.ylabel('Intensity (arb.)')
plt.show()
```
