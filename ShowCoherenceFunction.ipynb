### (2) Define a function that calculates the coherence function g(tau).

```python
# the index in the time vector is used for tau
def calcCoherenceFunction(U):
    
    # Calculate product U(t)*U(t+tau)
    def calcProduct(U,tau):
        av = 0;                       # summing variable
        lenVect = len(U)-tau          # stop early because finite vector length
        for n in range(lenVect):      # loop over all possible t values
            av = av + np.conj(U[n]) * U[n+tau] # calulate product and add to av                                               

        av = av/lenVect               # divide by number of data points  
        return av
    
    # main function
    ac = np.zeros(len(U),complex)        # create empty array 
    for j in range(len(U)):              # loop over all tau values
        ac[j] = calcProduct(U,j)         # product for each tau value 
    
    return ac/ac[0]

# calculate the coherence function
g = calcCoherenceFunction(U)
```
